Simple GAN

```python
import torch
import torch.nn as nn
```

```python
#Generator

class Generator(nn.Module):

    def __init__(self,noise_dim=100,img_channel=3):
        super().__init__()

        self.model = nn.Sequential(

            nn.Linear(noise_dim,256),
            nn.ReLU(True),
            nn.Linear(256,512),
            nn.ReLU(True),
            nn.Linear(512,64*64*img_channel),
            nn.Tanh()
        )

    def forward(self,inp_image):
        
        img = self.model(inp_image)
        img = img.view(img.size(0),3,64,64)
        return img
    

#discriminator 

class Discriminator(nn.Module):

    def __init__(self,inp_channel=3,):
        super().__init__()

        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(inp_channel*64*64,1024),
            nn.LeakyReLU(0.2,inplace=True),
            nn.Linear(1024,512),
            nn.LeakyReLU(0.2,inplace=True),
            nn.Linear(512,256),
            nn.LeakyReLU(),
            nn.Linear(256,1),
            nn.Sigmoid()
        )


    def forward(self,inp_image):
        return self.model(inp_image)

    

```
